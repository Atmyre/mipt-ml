### Модель:

Ну тут какая-то модель, к которой особо параметры не подбирались, уже не было времени. Это моя первая посылка:

```python
channel_axis = 1
freq_axis = 2
time_axis = 3
```

```python
def create_model():
    X_input = Input(shape=(1, 128, 431))
    # Input block
    x = BatchNormalization(axis=time_axis, name='bn_0_freq')(X_input)

    # Conv block 1
    x = Convolution2D(16, 1, 1, border_mode='same', name='conv1')(x)
    x = ELU()(x)
    x = MaxPooling2D(pool_size=(1, 1), name='pool1')(x)
    x = Dropout(0.1, name='dropout1')(x)

    # Conv block 2
    x = Convolution2D(32, 2, 2, border_mode='same', name='conv2')(x)
    x = ELU()(x)
    x = MaxPooling2D(pool_size=(, 2), name='pool2')(x)
    x = Dropout(0.1, name='dropout2')(x)

    # Conv block 3
    x = Convolution2D(64, 2, 2, border_mode='same', name='conv4')(x)
    x = ELU()(x)
    x = MaxPooling2D(pool_size=(1, 2), name='pool4')(x)
    x = Dropout(0.1, name='dropout3')(x)
    
    # Conv block 4
    x = Convolution2D(128, 3, 3, border_mode='same', name='conv5')(x)
    x = ELU()(x)
    x = MaxPooling2D(pool_size=(2, 2), name='pool5')(x)
    x = Dropout(0.1, name='dropout4')(x)

    # Output
    x = Flatten()(x)
    x = Dense(13, activation='sigmoid', name='output')(x)
    
    model = Model(X_input, x)
    return model
```

```python
model = create_model()
model.compile(optimizer='rmsprop',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

### Tренируем модель:

```python
%%time
model.fit(X, y)
```

Модель вернула вероятности всех 13 классов, которые мы предсказываем.Давайте выберем для кадждой мелодии наиболее вероятный класс и переведем его опять в строку (его название):

```python
def sort_result(tags, preds):
    result = zip(tags, preds)
    sorted_result = sorted(result, key=lambda x: x[1], reverse=True)
    return [(name, '%5.3f' % score) for name, score in sorted_result]

y_pred_names = []
for song_idx in range(X_test.shape[0]):
    sorted_result = sort_result(most_popular_genres, y_pred[song_idx, :].tolist())
    y_pred_names.append(sorted_result[0][0])
    print sorted_result
```
